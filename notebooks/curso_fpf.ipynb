```python
def listar_outliers(dataframe, coluna):
    Q1 = dataframe[coluna].quantile(0.25)
    Q3 = dataframe[coluna].quantile(0.75)
    IQR = Q3 - Q1
    
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    outliers = dataframe[(dataframe[coluna] < limite_inferior) | (dataframe[coluna] > limite_superior)]
    return outliers

outliers_encontrados = listar_outliers(df, 'trestbps')

print(f"Total de outliers encontrados: {len(outliers_encontrados)}")
display(outliers_encontrados.head())
```

# Usando IQR

```python
from scipy import stats

def detectar_outliers_iqr(df, coluna):
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inf = Q1 - 1.5 * IQR
    limite_sup = Q3 + 1.5 * IQR

    return (df[coluna] < limite_inf) | (df[coluna] > limite_sup)

def detectar_outliers_zscore(df, coluna, threshold=3):
    z_scores = np.abs(stats.zscore(df[coluna]))
    return z_scores > threshold

coluna_alvo = 'trestbps'

mask_outliers = detectar_outliers_iqr(df, coluna_alvo)
#mask_outliers = detectar_outliers_zscore(df, coluna_alvo)

df_somente_outliers = df[mask_outliers]
df_limpo = df[~mask_outliers]

print(f"--- RELATÓRIO: {coluna_alvo} ---")
print(f"Total Original: {len(df)}")
print(f"Total Removido: {len(df_somente_outliers)} ({len(df_somente_outliers)/len(df):.1%})")
print(f"Total Final:    {len(df_limpo)}")
print("\n--- Impacto na Estatística ---")
print(f"Média Antes:  {df[coluna_alvo].mean():.2f}")
print(f"Média Depois: {df_limpo[coluna_alvo].mean():.2f}")
print(f"Desvio Padrão Antes: {df[coluna_alvo].std():.2f}")
print(f"Desvio Padrão Depois: {df_limpo[coluna_alvo].std():.2f}")

df_limpo.to_csv('dataset_limpo.csv', index=False)
df_somente_outliers.to_csv('dataset_anomalias.csv', index=False)
print("\nArquivos salvos com sucesso!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from pmdarima import auto_arima
from sklearn.metrics import mean_absolute_error, mean_squared_error

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
df = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

plt.plot(df)
plt.title('Número de Passageiros (1949-1960)')
plt.ylabel('Passageiros')
plt.show()

missing = df.isnull().sum()
print(f"Dados faltantes:\n{missing}\n")

df_eda = df.copy()
df_eda['ano'] = df_eda.index.year
sns.boxplot(x='ano', y='Passengers', data=df_eda)
plt.title('Distribuição de Passageiros por Ano (Busca por Outliers)')
plt.show()

result = seasonal_decompose(df['Passengers'], model='multiplicative')
result.plot()
plt.show()
```

```python
df_prophet = df.reset_index()
df_prophet.columns = ['ds', 'y']


train_size = int(len(df) * 0.8)
train, test = df.iloc[:train_size], df.iloc[train_size:]

print(f"Tamanho do Treino: {len(train)} meses")
print(f"Tamanho do Teste: {len(test)} meses")

plt.plot(train.index, train['Passengers'], label='Treino')
plt.plot(test.index, test['Passengers'], label='Teste', color='orange')
plt.legend()
plt.title('Divisão Treino vs Teste')
plt.show()
```

```python
print("Treinando AutoARIMA...")

model = auto_arima(
    train['Passengers'],
    seasonal=True,
    m=12,
    trace=True,
    error_action='ignore',
    suppress_warnings=True,
    stepwise=True
)

print(f"\nMelhor modelo encontrado: {model.order} {model.seasonal_order}")

forecast = model.predict(n_periods=len(test))

forecast_series = pd.Series(forecast, index=test.index)
```

```python
mae = mean_absolute_error(test, forecast_series)
rmse = np.sqrt(mean_squared_error(test, forecast_series))
mape = np.mean(np.abs((test['Passengers'] - forecast_series) / test['Passengers'])) * 100

print(f"MAE: {mae:.2f} (Erro médio de {mae:.0f} passageiros)")
print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape:.2f}% (O erro percentual médio foi de {mape:.1f}%)")

plt.figure(figsize=(12,6))
plt.plot(train.index, train['Passengers'], label='Treino (Histórico)')
plt.plot(test.index, test['Passengers'], label='Real (Teste)')
plt.plot(forecast_series.index, forecast_series, label='Previsão (Modelo)', color='green', linestyle='--')
plt.title(f'Forecasting com SARIMA (MAPE: {mape:.2f}%)')
plt.legend()
plt.show()
```

```python
import xgboost as xgb
from pandas import concat

df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv',
                 parse_dates=['Month'], index_col='Month')

def create_features(df, label=None):
    """
    Cria features de tempo para o modelo de ML
    """
    df = df.copy()
    df['month'] = df.index.month
    
    df['lag_1'] = df['Passengers'].shift(1)
    df['lag_12'] = df['Passengers'].shift(12)
    
    df['rolling_mean_3'] = df['Passengers'].shift(1).rolling(window=3).mean()
    
    return df

df_ml = create_features(df)

df_ml = df_ml.dropna()

features = ['month', 'lag_1', 'lag_12', 'rolling_mean_3']
target = 'Passengers'

train_size = int(len(df_ml) * 0.8)
train_ml = df_ml.iloc[:train_size]
test_ml = df_ml.iloc[train_size:]

X_train = train_ml[features]
y_train = train_ml[target]
X_test = test_ml[features]
y_test = test_ml[target]

print(f"Features criadas:\n{X_train.head(3)}")
```

```python
reg = xgb.XGBRegressor(n_estimators=1000, learning_rate=0.01, early_stopping_rounds=50)

reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        verbose=False)

test_preds = []
current_batch = X_test.iloc[0].copy()

for i in range(len(test_ml)):
    pred = reg.predict(pd.DataFrame([current_batch]))[0]
    test_preds.append(pred)
    
    if i < len(test_ml) - 1:
        next_index = i + 1

        new_row = X_test.iloc[next_index].copy()
       
        new_row['lag_1'] = pred
        
        current_batch = new_row

mae_xgb = mean_absolute_error(y_test, test_preds)
print(f"XGBoost MAE: {mae_xgb:.2f}")

# Plot
plt.figure(figsize=(12,6))
plt.plot(test_ml.index, y_test, label='Real')
plt.plot(test_ml.index, test_preds, label='XGBoost Recursivo', color='purple')
plt.title(f'XGBoost Forecasting (MAE: {mae_xgb:.2f})')
plt.legend()
plt.show()
```

```python
from neuralforecast import NeuralForecast
from neuralforecast.models import NBEATS
from neuralforecast.losses.pytorch import MAE

df_dl = df.reset_index()
df_dl.columns = ['ds', 'y']
df_dl['unique_id'] = 'AirPassengers'

train_dl = df_dl.iloc[:train_size]
test_dl = df_dl.iloc[train_size:]

model_nbeats = NBEATS(
    input_size=12,              
    h=len(test_dl),             
    max_steps=500,              
    loss=MAE(),                 
    scaler_type='standard',    
    enable_progress_bar=True
)

nf = NeuralForecast(
    models=[model_nbeats],
    freq='MS'
)

forecast_df = nf.cross_validation(
    df=df_dl,
    val_size=len(test_dl),
    n_windows=1
)

y_true = forecast_df['y']
y_pred_nbeats = forecast_df['NBEATS']
mae_nbeats = mean_absolute_error(y_true, y_pred_nbeats)

print(f"N-BEATS MAE: {mae_nbeats:.2f}")

plt.figure(figsize=(12,6))
plt.plot(train_dl['ds'], train_dl['y'], label='Treino')
plt.plot(forecast_df['ds'], forecast_df['y'], label='Real')
plt.plot(forecast_df['ds'], forecast_df['NBEATS'], label='N-BEATS', color='red')
plt.title(f'N-BEATS Deep Learning (MAE: {mae_nbeats:.2f})')
plt.legend()
plt.show()
```
